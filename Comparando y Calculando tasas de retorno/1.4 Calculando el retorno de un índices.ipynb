```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as wb
```

### Cargando los datos

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI', '^BVSP']
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t]= wb.DataReader(t, data_source ='yahoo', start ='1997-1-1')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(20,5), title ='Normalizado')
plt.show()
```

```python
ind_retornos = (ind_data/ind_data.shift(1))-1
ind_retornos.tail()
```

```python
ind_ret_anual = ind_retornos.mean()*250
ind_ret_anual
```
